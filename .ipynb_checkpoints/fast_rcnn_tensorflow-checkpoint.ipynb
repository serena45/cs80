```python
# Usage of ROI Pooling op:

# output_features, argmax = roi_pooling_op(data, rois, output_shape)
# Inputs:
#     data: input feature tensor in NCHW (batch_size, num_channels, height, width) ordering
#     rois: input regions of interest of shape (num_batches, num_rois, 4) 
#           where the last dimension is (x, y, height, width) of the bounding-box
#     output_shape: tuple (output_height, output_height) of output feature map 
# Outputs:
#     output_features: output feature tensor of shape (batch_size, num_channels, num_rois, output_height, output_width)
#     argmax: flattened index into input feature tensor corresponding to each value in output_features. 
#             Same shape as output_features.
```

```python
# The width and height of the image
image_size = 227 # Must be divisible by the pooling layers

# Image depth
image_depth = 3

# The batch size
batch_size = 1

# number of classes
num_classes = 20
num_classes_with_background = num_classes + 1

# The number of region proposals
num_rois = 256

# The scaling factor for ROIs - equivalent to the combined downsampling of all prior strided accesses
# 4 (conv1) * 2 (pool1) * 2 (pool2) = 16
roi_scaling = 16

# Pixel means (probably for PASCAL VOC?) in BGR order
pixel_means = np.array([[[102.9801, 115.9465, 122.7717]]])
```

```python
# Helper functions for defining networks
def weight_variable(shape, wd):
    initial = tf.truncated_normal(shape, stddev=0.01)
    weight_decay = tf.mul(tf.nn.l2_loss(initial), wd)
    tf.add_to_collection('losses', weight_decay)
    return tf.Variable(initial)

def bias_variable(shape):
    initial = tf.constant(0.0, shape=shape)
    return tf.Variable(initial)

def conv2d(x, W, stride):
    return tf.nn.conv2d(x, W, strides=[1, stride, stride, 1], padding='SAME')

def conv2d_grouped(x, W, stride, group):
    # TODO wtf is going on here
    input_groups = tf.split(3, group, x)
    kernel_groups = tf.split(3, group, W)
    output_groups = [conv2d(i, k, stride) for i, k in zip(input_groups, kernel_groups)]
    return tf.concat(3, output_groups)

def max_pool(x, kernel_size, stride):
    return tf.nn.max_pool(x, ksize=[1, kernel_size, kernel_size, 1],
                        strides=[1, stride, stride, 1], padding='SAME')

def local_response_normalization(x):
    # TODO is 5 the correct radius?
    return tf.nn.local_response_normalization(x, depth_radius=5, alpha=.0001, beta=.75)
```

```python
# Something like Alexnet
sess = tf.InteractiveSession()

# TODO think about data ordering NCHW, etc
# TODO think about padding not matching up
# TODO look at graph output in visualizer to make sure it looks right
# TODO substitute in saved weights
# TODO fix the default initializations of weights and biases if we want to train

# TODO deal with ROI pooling layer taking NCHW
with tf.device("/cpu:0"):
    # Batched input
    input_data_tensor = tf.placeholder(tf.float32, shape=[None, image_size, image_size, image_depth], name="Input_Image_Batch") # batch size, image size, image size, image depth
    #y_ = tf.placeholder(tf.float32, shape=[None, num_classes], name="Input_Classes") # batch size, num_classes
    rois = tf.placeholder(tf.int32, shape=[None, num_rois, 4], name="Input_ROIs")

    # First Convolutional Layer
    # Variables
    W_conv1 = weight_variable([11, 11, image_depth, 96], wd=0.0) # filter size, filter size, input channels (image depth), output channels
    b_conv1 = bias_variable([96])
    # Layers
    conv1 = conv2d(input_data_tensor, W_conv1, stride=4)
    relu1 = tf.nn.relu(conv1 + b_conv1)
    pool1 = max_pool(relu1, kernel_size=3, stride=2)
    norm1 = local_response_normalization(pool1)

    # Second Convolutional Layer
    # Variables
    W_conv2 = weight_variable([5, 5, int(96 / 2), 256], wd=0.0) # filter size, filter size, input channels (image depth), output channels
    b_conv2 = bias_variable([256])
    # Layers
    #conv2 = conv2d(norm1, W_conv2, stride=1)
    conv2 = conv2d_grouped(norm1, W_conv2, stride=1, group=2)
    relu2 = tf.nn.relu(conv2 + b_conv2)
    pool2 = max_pool(relu2, kernel_size=3, stride=2)
    norm2 = local_response_normalization(pool2)

    # Third Convolutional Layer
    # Variables
    W_conv3 = weight_variable([3, 3, 256, 384], wd=0.0) # filter size, filter size, input channels (image depth), output channels
    b_conv3 = bias_variable([384])
    # Layers
    conv3 = conv2d(norm2, W_conv3, stride=1)
    relu3 = tf.nn.relu(conv3 + b_conv3)

    # Fourth Convolutional Layer
    # Variables
    # Variables
    W_conv4 = weight_variable([3, 3, int(384 / 2), 384], wd=0.0) # filter size, filter size, input channels (image depth), output channels
    b_conv4 = bias_variable([384])
    # Layers
    #conv4 = conv2d(relu3, W_conv4, stride=1)
    conv4 = conv2d_grouped(relu3, W_conv4, stride=1, group=2)
    relu4 = tf.nn.relu(conv4 + b_conv4)

    # Fifth Convolutional Layer
    # Variables
    W_conv5 = weight_variable([3, 3, int(384 / 2), 256], wd=0.0) # filter size, filter size, input channels (image depth), output channels
    b_conv5 = bias_variable([256])
    # Layers
    #conv5 = conv2d(relu4, W_conv5, stride=1)
    conv5 = conv2d_grouped(relu4, W_conv5, stride=1, group=2)
    relu5 = tf.nn.relu(conv5 + b_conv5)


    # ROI pooling to an output feature map of 6x6
    # TODO spatial scale
    # Convert NHWC to NCHW
    relu5_transpose = tf.transpose(relu5, [0, 3, 1, 2])
    output_dim_tensor = tf.constant((6,6))
    roi_pool5, argmax = roi_pooling_op(relu5_transpose, rois, output_dim_tensor)

    # TODO do we need to transpose again? 
    # ROI pooling outputs in NCRHW. Not sure what FC is expecting. 
    # Caffe layer appears to do NRCHW. Not sure about NR vs RN, but doesn't matter for single image.
    roi_pool5_transpose = tf.transpose(roi_pool5, [0, 2, 1, 3, 4])

    # Fully Connected 1
    # Weights
    # Weight variable is sized to match num_rois * roi_pooling_feature_length (6 x 6)
    W_fc6 = weight_variable([num_rois * 6 * 6, 4096], wd=0.0)
    b_fc6 = bias_variable([4096])
    # Layers
    roi_pool5_flat = tf.reshape(roi_pool5_transpose, [-1, 6 * 6 * num_rois])
    fc6 = tf.matmul(roi_pool5_flat, W_fc6)
    relu6 = tf.nn.relu(fc6 + b_fc6)

    # TODO dropout if we want to train

    # Fully Connected 2
    # Weights
    W_fc7 = weight_variable([4096, 4096], wd=0.004)
    b_fc7 = bias_variable([4096])
    # Layers
    fc7 = tf.matmul(relu6, W_fc7)
    relu7 = tf.nn.relu(fc7 + b_fc7)

    # TODO dropout if we want to train


    # Classification Score
    # Weights
    W_cls = weight_variable([4096, num_classes_with_background], wd=0.0)
    b_cls = bias_variable([num_classes_with_background])
    # Layers
    cls_score = tf.matmul(relu7, W_cls) + b_cls
    cls_prob=tf.nn.softmax(cls_score)

    # Bounding Box Regression Score
    # Weights
    W_bbox = weight_variable([4096, num_classes_with_background * 4], wd=0.0)
    b_bbox = bias_variable([num_classes_with_background * 4])
    # Layers
    bbox_pred = tf.matmul(relu7, W_bbox) + b_bbox
```

```python

candidates = set()
for r in regions:
    # excluding same rectangle (with different segments)
    if r['rect'] in candidates:
        continue
    # excluding regions smaller than 2000 pixels
    if r['size'] < 200:
        continue
    # distorted rects
    x, y, w, h = r['rect']
    if h == 0 or w==0:
        continue
    if w / h > 4 or h / w > 4:
        continue
    candidates.add((x, y, h, w))
```

```python
def vis_detections(im, class_name, dets, thresh=0.5):
    """Draw detected bounding boxes."""
    inds = np.where(dets[:, -1] >= thresh)[0]
    if len(inds) == 0:
        return

    im = im[:, :, (2, 1, 0)]
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.imshow(im, aspect='equal')
    for i in inds:
        print("sample in category {}".format(class_name))
        bbox = dets[i, :4]
        score = dets[i, -1]

        ax.add_patch(
            plt.Rectangle((bbox[0], bbox[1]),
                          bbox[2] - bbox[0],
                          bbox[3] - bbox[1], fill=False,
                          edgecolor='red', linewidth=3.5)
            )
        ax.text(bbox[0], bbox[1] - 2,
                '{:s} {:.3f}'.format(class_name, score),
                bbox=dict(facecolor='blue', alpha=0.5),
                fontsize=14, color='white')

    ax.set_title(('{} detections with '
                  'p({} | box) >= {:.1f}').format(class_name, class_name,
                                                  thresh),
                  fontsize=14)
    plt.axis('off')
    plt.tight_layout()
    plt.draw()
```

```python
def nms(dets, thresh):
    x1 = dets[:, 0]
    y1 = dets[:, 1]
    x2 = dets[:, 2]
    y2 = dets[:, 3]
    scores = dets[:, 4]

    areas = (x2 - x1 + 1) * (y2 - y1 + 1)
    order = scores.argsort()[::-1]

    keep = []
    while order.size > 0:
        i = order[0]
        keep.append(i)
        xx1 = np.maximum(x1[i], x1[order[1:]])
        yy1 = np.maximum(y1[i], y1[order[1:]])
        xx2 = np.minimum(x2[i], x2[order[1:]])
        yy2 = np.minimum(y2[i], y2[order[1:]])

        w = np.maximum(0.0, xx2 - xx1 + 1)
        h = np.maximum(0.0, yy2 - yy1 + 1)
        inter = w * h
        ovr = inter / (areas[i] + areas[order[1:]] - inter)

        inds = np.where(ovr <= thresh)[0]
        order = order[inds + 1]

    return keep
```

```python
im = img
scores = cls_out
boxes = bbox_out

# Visualize detections for each class
CONF_THRESH = 0.1
NMS_THRESH = 0.1
for cls in classes:
    cls_ind = CLASSES.index(cls)
    cls_boxes = boxes[:, 4*cls_ind:4*(cls_ind + 1)]
    cls_scores = scores[:, cls_ind]
    keep = np.where(cls_scores >= CONF_THRESH)[0]
    cls_boxes = cls_boxes[keep, :]
    cls_scores = cls_scores[keep]
    dets = np.hstack((cls_boxes,
                      cls_scores[:, np.newaxis])).astype(np.float32)
    keep = nms(dets, NMS_THRESH)
    dets = dets[keep, :]
    print('All {} detections with p({} | box) >= {:.1f}'.format(cls, cls, \
                                                                CONF_THRESH))
    vis_detections(im, cls, dets, thresh=CONF_THRESH)
```
